# Repair effectiveness (precision / recall / F1)

Scores a Horizon repair run against ground truth, per paper §6.1. Compares three
row-aligned tables:

- **clean** — ground truth (`clean.csv`)
- **dirty** — the injected errors fed to Horizon (`injected/<error>.csv`)
- **cleaned** — Horizon's repair output (`output/<dataset>_cleaned_data.csv`)

All three are read as `Utf8` so values compare like-for-like. The single call to
`evaluate_repair` does the cell-level counting and calls `precision`, `recall`, `f1`.

```python
import sys
from pathlib import Path

import polars as pl

# project root on sys.path so `eval` resolves from notebooks/
sys.path.insert(0, str(Path("..").resolve()))

from eval.effectiveness_eval import evaluate_repair
```

```python
dataset_dir = Path("..") / "datasets" / "insurance_claims_58k"

clean_path = dataset_dir / "clean.csv"
dirty_path = dataset_dir / "injected" / "e2_r10.csv"
cleaned_path = Path("..") / "output" / "insurance_claims_58k_cleaned_data.csv"
```

```python
# read as Utf8 (infer_schema_length=0) so the three tables compare like-for-like
clean = pl.read_csv(clean_path, infer_schema_length=0)
dirty = pl.read_csv(dirty_path, infer_schema_length=0)
cleaned = pl.read_csv(cleaned_path, infer_schema_length=0)


print(f"clean   {clean.shape}")
print(f"dirty   {dirty.shape}")
print(f"cleaned {cleaned.shape}")
```

```python
from pprint import pprint

pprint(evaluate_repair(clean, dirty, cleaned), sort_dicts=False)
```
