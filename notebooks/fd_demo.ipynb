# FunctionalDependency

Demonstrates the `FunctionalDependency` class and the `get_fds` loader using the beers dataset.

```python
import sys
from pathlib import Path

# add project root to path so horizon package is importable from notebooks/
sys.path.insert(0, str(Path("..").resolve()))

from horizon.fd import FunctionalDependency
from horizon.loaders import get_fds
```

## Loading FDs from CSV

```python
fds_path = Path("..") / "datasets" / "beers" / "fds.csv"
fds = get_fds(fds_path)

print(f"Loaded {len(fds)} functional dependencies")
fds
```

## Accessing LHS and RHS

```python
fd = fds[0]
print(f"LHS: {fd.lhs}")
print(f"RHS: {fd.rhs}")
```

## Composite LHS

Some FDs have multiple attributes on the left-hand side.

```python
composite = [fd for fd in fds if len(fd.lhs) > 1]
print(f"{len(composite)} FDs with composite LHS:")
composite
```

## Equality and use in sets

```python
fd_a = FunctionalDependency(["beer-name", "brewery-name"], "brewery_id")
fd_b = FunctionalDependency(("beer-name", "brewery-name"), "brewery_id")

print(f"Equal:        {fd_a == fd_b}")
print(f"In loaded set: {fd_a in set(fds)}")
```
